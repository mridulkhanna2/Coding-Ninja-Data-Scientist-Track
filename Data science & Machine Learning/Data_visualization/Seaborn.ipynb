### 2 a) Categorical scatter plots: 

### stripplot()

```python
df=sns.load_dataset('tips')
```

```python
sns.catplot(x='sex',y='total_bill',data=df)
plt.show()
```

```python
sns.catplot(x='sex',y='total_bill',data=df,jitter=0.25)
plt.show()
#jitter helps us control the randomness of these points
```

```python
sns.catplot(x='sex',y='total_bill',data=df,jitter=0.75)
plt.show()
#jitter helps us control the randomness of these points
```

```python
sns.catplot(x='sex',y='total_bill',data=df,jitter=0.005)
plt.show()
#jitter helps us control the randomness of these points
```

multople points are overlapping on each in the above
swarmplot does not allow overlapping by using a hidden algoritthm

### Swarmplot()
<b> generally works well for small data </b>

```python
sns.catplot(x='sex',y='total_bill',data=df,kind='swarm')
plt.show()
#jitter helps us control the randomness of these points
```

```python
sns.catplot(x='sex',y='total_bill',data=df,kind='swarm',hue='smoker')
plt.show()
#jitter helps us control the randomness of these points
```

```python
sns.catplot(x='sex',y='total_bill',data=df,hue='smoker',jitter=0.005)
plt.show()
'''very difficult tp understand how values are distributed in terms of 
smoker column so here swarmplot is better'''
```
